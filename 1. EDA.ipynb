```python
df = df1.merge(df2, on=['testID', 'assessmentItemID'], how='left')
df = df.merge(df3, on=['learnerID', 'testID'], how='left')
# 데이터 병합하기

print(df.head())
print(df.info())
```

```python
# EDA

# 학생 점수 분포 시각화
sns.histplot(df['answerCode'], bins=2, kde=False)
plt.title('Distribution of Scores (Correct: 1, Incorrect: 0)')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()

# 문항 난이도 분포 시각화
sns.histplot(df['difficultyLevel'])
plt.title('Distribution of Difficulty Levels')
plt.xlabel('Difficulty Level')
plt.ylabel('Count')
plt.show()

# 변별도 분포 시각화
sns.histplot(df['discriminationLevel'])
plt.title('Distribution of Discrimination Levels')
plt.xlabel('Discrimination Level')
plt.ylabel('Count')
plt.show()

# 추측도 분포 시각화
sns.histplot(df['guessLevel'], bins=10)
plt.title('Distribution of Guess Levels')
plt.xlabel('Guess Level')
plt.ylabel('Count')
plt.show()
```

```python
# 변별도 분포 그래프가 이상하다
# 구간을 30개로 나누어서 다시 그려보기

sns.histplot(df['discriminationLevel'], bins=30)
plt.title('Original Distribution of Discrimination Levels')
plt.xlabel('Discrimination Level')
plt.ylabel('Count')
plt.show()
```

```python
# 기술 통계량 확인
print(df['discriminationLevel'].describe())

# 고유 값과 빈도 확인
print(df['discriminationLevel'].value_counts())
```

```python
# 최대값이 4000대인 반면, 표준편차는 73 가량이고, 75퍼센트까지가 3.24인걸로 보아
# 극단적으로 큰 값들이 포함되어 보인다.
# -> 로그변환을 시도해봐야겠다.

df['log_discriminationLevel'] = np.log1p(df['discriminationLevel'])


# 변환 후 다시 히스토그램 시각화
sns.histplot(df['log_discriminationLevel'], bins=30)
plt.title('Log-Transformed Distribution of Discrimination Levels')
plt.xlabel('Log Discrimination Level')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# 로그 변환된 데이터 히스토그램
sns.histplot(df['log_discriminationLevel'], bins=30, kde=True)
plt.title('Log-Transformed Distribution of Discrimination Levels')
plt.xlabel('Log Discrimination Level')
plt.ylabel('Count')
plt.show()

# 로그 변환된 데이터 박스 플롯
sns.boxplot(x=df['log_discriminationLevel'])
plt.title('Boxplot of Log-Transformed Discrimination Levels')
plt.xlabel('Log Discrimination Level')
plt.show()

# 로그 변환된 데이터 Q-Q 플롯
stats.probplot(df['log_discriminationLevel'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Log-Transformed Discrimination Levels')
plt.show()
```

```python
# 왜도와 첨도 계산
skewness = df['log_discriminationLevel'].skew()
kurtosis = df['log_discriminationLevel'].kurtosis()

print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurtosis}')
```

왜도(Skewness)
<pre>
왜도의 절대값이 1 이상이면 데이터가 상당히 비대칭
왜도의 절대값이 0.5 이상이면 데이터가 조금 비대칭
왜도의 절대값이 0.5 미만이면 데이터가 비교적 대칭
</pre>
첨도(Kurtosis)
<pre>
첨도가 3에 가까우면 정규 분포와 비슷
첨도가 3보다 크면 데이터가 뾰족한 꼬리. (양수 첨도)
첨도가 3보다 작으면 데이터가 둥글고 분산이 작음 (음수 첨도)
</pre>
